```python
import pandas as p
import numpy as n
```

```python
d = p.read_csv('AB_NYC_2019.csv')
d
```

```python
d = d.drop_duplicates()
d
```

```python
d = d.drop(columns = ['name', 'neighbourhood', 'calculated_host_listings_count'])
d.head(5)
```

```python
d.info()
```

```python
d['host_name'] = d['host_name'].astype(str)
d['host_name'] = d['host_name'].str.replace("nan", " ")
d['host_name'] = d['host_name'].str.replace('[^a-zA-Z]', '',regex = True)
d['host_name'] = d['host_name'].str.replace(r'([a-z])([A-Z])',r'\1 \2', regex = True)
d.head(5)
```

```python
d_priced = d[d['price'] != 0]
d_priced.head(5)
```

```python
d_year = d[d['minimum_nights'] <= 367]
d_year.head(5)
```

```python
d_avail = d[d['availability_365'] != 0]
d_avail.head(5)
```

```python
d['last_review'] = d['last_review'].astype(str)
d['last_review'] = d['last_review'].str.replace("nan", " ")
d.head(5)
```
