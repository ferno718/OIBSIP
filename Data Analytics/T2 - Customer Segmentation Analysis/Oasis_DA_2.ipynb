```python
import pandas as p
import numpy as n
import matplotlib.pyplot as m
import seaborn as s
```

```python
d = p.read_csv('ifood_df.csv')
d.head()
```

```python
d.drop_duplicates()
d.head()
```

```python
d.nunique()
```

```python
d.drop(columns = ['Z_CostContact', 'Z_Revenue'])
```

```python
d.describe()
```

```python
def get_relationship(row):
  if row['marital_Married'] == 1:
    return 1
  elif row['marital_Together'] == 1:
    return 1
  else:
    return 0

d['In_relationship'] = d.apply(get_relationship, axis = 1)
d.head()
```

```python
d.rename(columns = {'education_2n Cycle\t': 'education_2n Cycle'}, inplace = True)
```

```python
def Education_status(row):
  if row['education_2n Cycle'] == 1:
    return '2'
  elif row['education_Basic'] == 1:
      return '1'
  elif row['education_Graduation'] == 1:
      return '3'
  elif row['education_Master'] == 1:
      return '4'
  elif row['education_PhD'] == 1:
      return '5'
  else:
      return 'Unknown'

d['Education_status'] = d.apply(Education_status, axis = 1)
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
cols_to_cluster = ['Income', 'MntTotal', 'In_relationship', 'Education_status']
d_scaled = d.copy()
d_scaled[cols_to_cluster] = scaler.fit_transform(d[cols_to_cluster])
```

```python
x = d_scaled[cols_to_cluster]
inertia_list = []
for k in range(1,11):
  inertia = KMeans(n_clusters = k, n_init = 10).fit(x).inertia_
  inertia_list.append(inertia)
```

```python
m.figure(figsize = (7,4))
m.plot(range(1,11), inertia_list, color = 'c', marker = 'd', mfc = 'pink', mec = 'k')
m.title("Inertia vs No. of Clusters")
m.xlabel("No. of Clusters (K)")
m.ylabel("Inertia")
m.show()
```

```python
model = KMeans(n_clusters = 4, random_state = 7)
model.fit(d_scaled[cols_to_cluster])
d_scaled['Cluster'] = model.predict(d_scaled[cols_to_cluster])
```

```python
from sklearn import decomposition
pca = decomposition.PCA(n_components = 2)
pca_res = pca.fit_transform(d_scaled[cols_to_cluster])
d_scaled['pc1'] = pca_res[:,0]
d_scaled['pc2'] = pca_res[:,1]
```

```python
m.figure(figsize = (9,5))
s.scatterplot(x = 'pc1', y = 'pc2', data = d_scaled, hue = 'Cluster', palette = 'viridis')
m.title("Cluster Visualization")
m.xlabel("Principal Component 1 (pc1)")
m.ylabel("Principal Component 2 (pc2)")
m.legend(title = 'Clusters')
```

```python
cnum = d_scaled.groupby('Cluster')['MntTotal'].count().reset_index()

m.figure(figsize = (9,7))
s.barplot(x = 'Cluster', y = 'MntTotal', data = cnum, palette = 'viridis')
m.title('Cluster sizes')
m.xlabel('Cluster')
m.ylabel('Cluster_num')
m.show()
```

```python
scaler = StandardScaler()
cols_to_cluster = ['Income', 'MntTotal', 'In_relationship']
d_clus = d.copy()

model = KMeans(n_clusters = 4, random_state = 7)
model.fit(d_clus[cols_to_cluster])
d_clus['Cluster'] = model.predict(d_clus[cols_to_cluster])
d_clus.head()
```
