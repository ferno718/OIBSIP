**Importing necessary libraries**

```python
import matplotlib.pyplot as m
import numpy as n
import pandas as p
import seaborn as s
```



---


**Reading the csv file**

---



```python
d = p.read_csv('Unemployment.csv')
```



---


**Data Cleaning**

---



```python
d.isnull().sum()
```

```python
d.duplicated().sum()
```

```python
d = d.dropna()
```

```python
d.columns = ['Region','Date','Frequency','Estimated Unemployment Rate','Estimated Employed','Estimated Labour Participation Rate','Area']
```

```python
d['Estimated Unemployement Rate'] = p.to_numeric(d['Estimated Employed'], errors='coerce')
d['Estimated Employed'] = p.to_numeric(d['Estimated Employed'], errors='coerce')
d['Estimated Labour Participation Rate'] = p.to_numeric(d['Estimated Employed'], errors='coerce')
d['Date'] = p.to_datetime(d['Date'])
```

**Data Visualization**


---



---

*   ***Unemployment Rate (State-Wise) by Bar Graph***

---

```python
d1 = d[['Region','Estimated Unemployment Rate']].groupby('Region').sum().sort_values('Estimated Unemployment Rate', ascending = False).head(10)
```

```python
m.figure(figsize = [15,7])
s.barplot(x = 'Region', y = 'Estimated Unemployment Rate', data = d, palette = 'flare', hue = 'Region')
m.title('Unemployment Rate (State-Wise) by Bar Graph')
m.xlabel('Region')
m.ylabel('Unemployment Rate (%)')
m.xticks(rotation=85)
m.show()
```

---

*   ***Top 10 States with Highest Unemployment Rate***

---

```python
m.figure(figsize=(12, 6))
s.barplot(x = 'Region', y = 'Estimated Unemployment Rate', data = d1, palette = "crest", hue = "Region")
m.xlabel('Region')
m.ylabel('Estimated Unemployment Rate')
m.title('Top 10 Regions by Estimated Unemployment Rate')
m.xticks(rotation = 45)
m.show()
```



---

*   ***Estimated Employment Rate by Line Graph***

---



```python
y = d['Estimated Employed']
m.figure(figsize = [18,9])
m.plot(y, label = 'Sales', marker = 'o', mec = 'purple', mfc = 'pink', color = 'c', markersize = 2)
m.title('Estimated Employment Rate by Line Graph')
m.xlabel('Index')
m.ylabel('Estimated Employed')
m.legend(facecolor = 'lightgrey', edgecolor = 'k', shadow = True)
m.xticks(rotation = 85)
m.show()
```



---

*   ***Distribution of Estimated Employement***

---



```python
m.figure(figsize = (12, 6))
s.histplot(data = d, x = 'Estimated Employed', bins = 30, kde = True, color = 'r')
m.xlabel('Estimated Employed')
m.ylabel('Frequency')
m.title('Distribution of Estimated Employed')
m.show()
```



---

*   ***Kernel Density Estimation of Estimated Employed***

---



```python
m.figure(figsize=(12, 6))
s.kdeplot(data = d, x = 'Estimated Employed', color = 'hotpink', fill = True)
m.xlabel('Estimated Employed')
m.ylabel('Density')
m.title('Kernel Density Estimation of Estimated Employed')
m.show()
```



---

*   ***Box Plot of Estimated Employed***

---



```python
m.figure(figsize=(15,7))
s.boxplot(data = d, x = 'Region', y = 'Estimated Employed', palette = 'rocket_r', hue = 'Region')
m.ylabel('Region')
m.ylabel('Estimated Employed')
m.title('Box Plot of Estimated Employed')
m.xticks(rotation = 85)
m.show()
```



---

*   ***Joint Plot with Regression for Estimated Employed vs. Estimated Unemployment Rate***

---



```python
s.set(style="darkgrid")
s.jointplot(x = 'Estimated Employed', y = 'Estimated Unemployment Rate', data = d, kind = 'reg', height = 7, color = 'cyan', joint_kws = {'color' : 'k'})
m.suptitle('Joint Plot with Regression for Estimated Employed vs. Estimated Unemployment Rate', y = 1.02)
m.show()
```



---

*   ***Joint Plot with Hexbin for Estimated Labour Participation Rate vs. Estimated Unemployment Rate***

---



```python
s.set(style="darkgrid")
s.jointplot(x = 'Estimated Labour Participation Rate', y = 'Estimated Unemployment Rate', data = d, kind = 'hex', color = 'k', height = 7, joint_kws = {'color' : 'crimson'})
m.suptitle('Joint Plot with Hexbin for Estimated Labour Participation Rate vs. Estimated Unemployment Rate', y=1.02)
m.show()
```



---

*   ***Joint Plot with KDE and Marginal Histograms for Estimated Unemployment Rate vs. Estimated Labour Participation Rate***

---



```python
s.set(style="darkgrid")
g = s.jointplot(x='Estimated Unemployment Rate', y='Estimated Labour Participation Rate', data=d, kind='kde', color='g', fill=True, height=7)
g.plot_joint(s.scatterplot, marker='.', color='k')
g.ax_marg_x.hist(d['Estimated Unemployment Rate'], color='k', alpha=0.6, bins=40)
g.ax_marg_y.hist(d['Estimated Labour Participation Rate'], color='k', alpha=0.6, orientation="horizontal", bins=40)
m.suptitle('Joint Plot with KDE and Marginal Histograms for Estimated Unemployment Rate vs. Estimated Labour Participation Rate', y=1.02)
m.show()
```




---

*   ***Pairplot of Estimated Unemployment Rate, Estimated Employed, and Estimated Labour Participation Rate***

---



```python
s.set(style="darkgrid")
m.figure(figsize = [15,7])
s.pairplot(d[['Region', 'Frequency', 'Estimated Unemployment Rate', 'Estimated Employed', 'Estimated Labour Participation Rate']], palette = 'magma', hue = 'Region', height = 3.18)
m.suptitle('Pairplot of Estimated Unemployment Rate, Estimated Employed, and Estimated Labour Participation Rate', y=1.02)
m.show()
```
